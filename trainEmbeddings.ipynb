```python
COLUMNS_USED = ['desc', 'title', 'grade']
TEXT_FEATURES = ['desc', 'title']

df = df.loc[:, COLUMNS_USED]
df = df.reset_index(drop=True)
df = df.dropna(axis = 0)
df.shape
```

```python
df[TEXT_FEATURES] = df[TEXT_FEATURES].replace(regex=True, 
                                              to_replace=r'^.+ added on [^A-za-z]+ > ', value=r'')
df[TEXT_FEATURES] = df[TEXT_FEATURES].replace(regex=True, 
                                              to_replace=r'<br/>', value=r'')
```

```python
# train/dev/test split is 80/10/10%
train_set, test_and_dev_set = train_test_split(df, test_size=0.20)
dev_set, test_set = train_test_split(test_and_dev_set, test_size=0.5)
print(train_set.shape)
print(dev_set.shape)
print(test_set.shape)
```

```python
sentences = []
for text in (train_set['desc'] + train_set['title']):
    for sentence in gensim.summarization.textcleaner.get_sentences(text):
        sentence = sentence.lower()
        sentence = re.sub(r'[^a-z ]', ' ', sentence)
        sentences.append(sentence.split())
```

```python
model = Word2Vec(sentences, size=50)
```

```python
processed_descs = []
processed_titles = []

for index, row in df.iterrows():
    print("dbg", index)
    desc_vector = np.zeros(model['credit'].shape)
    title_vector = np.zeros(model['credit'].shape)
    desc = re.sub(r'[^a-z ]', ' ', row['desc'].lower()).split()
    title = re.sub(r'[^a-z ]', ' ', row['title'].lower()).split()
    for word in desc:
        if word in model:
            desc_vector += model[word]
    desc_vector /= len(desc)
    for word in title:
        if word in model:
            title_vector += model[word]
    title_vector /= len(title)
    processed_descs.append(desc_vector)
    processed_titles.append(title_vector)
```
